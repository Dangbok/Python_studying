## dataframe selection

### row(s)

#### loc

```python
print("## one row")
df1=pd.DataFrame(np.arange(16).reshape(4,4), index=[10,20,30,40], columns=list('abcd'), dtype=float)
df1
df1.loc[10,:]
df1.loc[10]

print("## one row (dtype 주의)")
df1['c']=df1['c'].astype(str)
df1.loc[10]
type(df1.loc[10]['a'])
type(df1.loc[10]['c'])

print("## rows")
df1.loc[10:30] # 경계포함 O
df1.loc[10:30].dtypes

print("## rows (fancy index)")
df1.loc[[10,20,40]]
df1.loc[[40,20,10]]
```

#### iloc

```python
dic1={'a':[1,2,3],'b':[2,3,4], 'c':[3,4,5]}
df1=pd.DataFrame(data=dic1, index=list('gkp'))
df1
df1.iloc[2]
df1.iloc[0:2] # 경계 포함 X
df1.iloc[[2]] # dataframe return (not series)
df1.iloc[[0,2]] # dataframe return

## head & tail
d1=pd.read_table('pandas_data1.txt', sep='\t', encoding='cp949', index_col='이름')
d1.index.name=None

d1.head(4)
d1.iloc[:4]

d1.tail()
d1.iloc[-5:]

## head for series
d1['기말'].head()
```

## dataframe boolean selection

### boolean 값 직접 입력

```python
df1=pd.DataFrame(np.arange(16).reshape(4,4), index=[10,20,30,40], columns=list('abcd'), dtype=float)

print("## loc")
df1.loc[[True, True, False, True],['a','b']]
df1.loc[20, [True, False, False, True]]

print("## iloc")
df1.iloc[[True, True, False, True],[0,2]]
df1.iloc[1, [True, False, False, True]]
```

### 조건 검사 결과 입력

```python
df1=pd.DataFrame(np.arange(16).reshape(4,4), index=[10,20,30,40], columns=list('abcd'), dtype=float)

## 조건 검사
df1
df1['c']>9

## 조건 검사 결과 입력

df1.loc[df1['c']>9]

df1.loc[df1['c']>9,'d']

df1.loc[df1['c']>9,['d']]

df1.loc[df1['c']>9,['c','d']]
```

```python
## 1
# m1 (method 1)
np.sum(df1['final']>=40)
# m2
df2=df1.loc[df1['final']>=40]
len(df2)
```

```python
## 3
s1=df1.loc[df1['att']<=5,'final']
np.mean(s1)
```

```python
## 4
df1.loc[df1['proj']==100,['mid','final','att']]
```
