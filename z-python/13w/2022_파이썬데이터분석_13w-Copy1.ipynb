### dataframe에 적용해보기

```python
def att_pass_or_fail(x):
    if x>5:
        return 'P'
    else:
        return 'F'

def proj_norm(x):
    return x*25/100

d1=pd.read_table('pandas_data3.txt', index_col=0)
d1=d1.head()
d1
d1['att_PF1']=d1['att'].map(att_pass_or_fail)
d1['att_PF2']=np.where(d1['att']>5, 'P', 'F')
d1

d1['proj_norm1']=d1['proj'].map(proj_norm)
d1['proj_norm2']=d1['proj']*25/100
d1['proj_norm3']=d1['proj'].map(lambda x: x*25/100)
d1

d1['gender']=d1['gender'].replace({'F':1,'M':0})
d1
```

### 실습2

```python
def get_mid_grade(x):
    if x>=25:
        return 'A'
    elif x>=20:
        return 'B'
    else:
        return 'C'

def is_senior(x):
    if x in [3,4]:
        return True
    else:
        return False

def get_mid_class(x):
    global mid_mean
    if x>mid_mean:
        return 'H'
    else:
        return 'L'


d1=pd.read_table('pandas_data3.txt', index_col=0)
d1=d1.head(10)

# 1
d1['final_norm']=d1['final']*0.5

# 2
d1['mid_grade']=d1['mid'].map(get_mid_grade)

# 3
d1['senior']=d1['year'].map(is_senior)

# 4
d1['year']=d1['year'].map({1:'first',2:'second',3:'third',4:'fourth'})

# 5
mid_mean=d1['mid'].mean()
mid_mean
d1['mid_HL']=d1['mid'].map(get_mid_class)
d1
```

### 이미 정의되어있는 함수 사용 가능

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:3,1:]
d2

def f2(x):
    return x.mean()

d2.apply(f2, axis=1)
d2.apply(lambda x: x.mean(), axis=1)
d2.apply(np.mean, axis=1)
d2.mean(axis=1)
```

### apply에 복잡한 함수 사용

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:5,1:]
d2

def f2(x):
    x2=x.loc[x%2==0]
        
    if np.sum(x2)%3==0:
        return 'G'
    else:
        return 'B'

d2.loc['G/B']=d2.apply(f2)
d2

```

### apply 적용 결과를 새로운 column row에 추가

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:5,1:]
d2

def f3(x):
    return np.mean(x)+np.std(x)

d2['num_of_3mul']=d2.apply(lambda x: np.sum(x%3==0), axis=1)
d2

d2.loc['temp']=d2.apply(f3)
d2
```

### apply with return series 함수

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:5,1:]
d2

def f3(x):
    v1=np.sum(x)
    v2=np.mean(x)
    v3=np.std(x)
    
    return pd.Series([v1,v2,v3], index=['sum','mean','std'])

d2.apply(f3)
d2.apply(f3, axis=1)

d3=d2.apply(f3)
pd.concat([d2,d3], axis=0)
```

### 실습3

```python
# 1)
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:10,1:]
d2

def f1(col):
    m1=np.mean(col)
    s1=np.std(col)
    b1=((m1-s1)<col)&(col<(m1+s1))
    return np.sum(col.loc[b1])

d2.apply(f1)
```

```python
# 2)
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:10,1:]
d2

def f2(x):
    if x>=10:
        return 'A'
    if x%2==0:
        return 'B'
    return 'C'
    
d2.applymap(f2)
```

```python
# 3)
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:10,1:]
d2

def f3(row):
    v1=row.loc['mid']+row.loc['final']+row.loc['proj']
    if v1>=140:
        return 'P'
    return 'F'
    
d2['P_or_F']=d2.apply(f3, axis=1)
d2
```

```python
# 4)
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:10,1:]
d2

def f4(col):
    m1=np.mean(col)
    s1=np.std(col)
    x1=np.max(col)
    n1=np.min(col)
    return pd.Series([m1,s1,x1,n1], index=['mean','std','max','min'])

d3=d2.apply(f4)
pd.concat([d2,d3])
```

```python
# 5)
d1=pd.read_table('pandas_data3.txt', index_col=0)
d2=d1.iloc[:10,1:]
d2

mid_mean=d2['mid'].mean()
fin_mean=d2['final'].mean()
mid_mean
fin_mean

def f5(row):
    global mid_mean, fin_mean
    if (row['mid']>=mid_mean) or (row['final']>=fin_mean):
        return 'P'
    return 'F'

d2['P_or_F']=d2.apply(f5,axis=1)
d2
```
