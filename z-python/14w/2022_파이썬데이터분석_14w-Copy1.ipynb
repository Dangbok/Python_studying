## merge

### 연습

```python
merge join concat
```

```python
df1 = pd.DataFrame({'key1': ['a', 'b', 'c', 'd'], 'v1': [1, 2, 3, 4], 'v2': [5, 6, 7, 8]})
df2 = pd.DataFrame({'key2': ['a', 'd', 'c', 'b'], 'v3': [1, 2, 3, 4], 'v4': [5, 6, 7, 8]})
df1
df2

pd.merge(df1,df2, left_on='key1', right_on='key2')
```

```python
# merge 기준 column key1, key2에 다른 element가 있는 경우
df1 = pd.DataFrame({'key1': ['a', 'b', 'c', 'e'], 'v1': [1, 2, 3, 4], 'v2': [5, 6, 7, 8]})
df2 = pd.DataFrame({'key2': ['a', 'd', 'c', 'b'], 'v3': [1, 2, 3, 4], 'v4': [5, 6, 7, 8]})
df1
df2

pd.merge(df1,df2, left_on='key1', right_on='key2')
pd.merge(df1,df2, left_on='key1', right_on='key2', how='outer')
pd.merge(df1,df2, left_on='key1', right_on='key2', how='left')
```

```python
# merge 기준 column이름이 key로 같은 경우
df1 = pd.DataFrame({'key': ['a', 'b', 'c', 'e'], 'v1': [1, 2, 3, 4], 'v2': [5, 6, 7, 8]})
df2 = pd.DataFrame({'key': ['a', 'd', 'c', 'b'], 'v3': [1, 2, 3, 4], 'v4': [5, 6, 7, 8]})
df1
df2

pd.merge(df1,df2, left_on='key', right_on='key')
```

```python
# merge가 되는 column이름에 같은 이름 (v1)이 있는 경우
df1 = pd.DataFrame({'key': ['a', 'b', 'c', 'e'], 'v1': [1, 2, 3, 4], 'v2': [5, 6, 7, 8]})
df2 = pd.DataFrame({'key': ['a', 'd', 'c', 'b'], 'v1': [1, 2, 3, 4], 'v4': [5, 6, 7, 8]})
df1
df2

pd.merge(df1,df2, left_on='key', right_on='key')
```

```python
# merge 기준을 index로 하는 경우
df1 = pd.DataFrame({'key1': ['a', 'b', 'c', 'e'], 'v1': [1, 2, 3, 4], 'v2': [5, 6, 7, 8]}).set_index('key1')
df2 = pd.DataFrame({'key2': ['a', 'd', 'c', 'b'], 'v3': [1, 2, 3, 4], 'v4': [5, 6, 7, 8]}).set_index('key2')
df1
df2

pd.merge(df1,df2, left_index=True, right_index=True)
pd.merge(df1,df2, left_index=True, right_index=True, how='outer')
```
