```python
d1=pd.read_table('pandas_data3-Copy11.txt',index_col='name',sep='\t',encoding='cp949')
d2=d1.iloc[:10,1:]
d2

def f(col):
    m1=np.mean(col)
    s1=np.std(col)
    x1=np.max(col)
    n1=np.min(col)
    return pd.Series([m1,s1,x1,n1],index=['mean','std','max','min'])

d3=d2.apply(f)
d3
pd.concat([d2,d3])
```

```python
d1=pd.read_table('pandas_data3-Copy11.txt',index_col='name',sep='\t',encoding='cp949')
d2=d1.iloc[:10,1:]
d2


mid_mean=d2['mid'].mean()
final_mean=d2['final'].mean()

def f(row):
    global final_mean,mid_mean
    if (row['mid']>=mid_mean)or(row['final']>=final_mean):
        return 'P'
    return 'F'

d2['PF']=d2.apply(f,axis=1)
d2
```
