## 과제1

```python
## 주어진 d1에 final_grade column을 추가하는데,
## final_grade column에는 "final평균 + final표준편차" 보다 final점수가 큰 경우에는 A, 30점 이하는 C, 그 사이는 B를 넣으시오

d1=pd.read_table('pandas_data3-Copy1.txt', index_col=0)
d1=d1.head(10)
d1

# 코드 작성 시작

def final_grade(x):
    global final_mean,final_std
    if x>final_mean+final_std:
        return 'A'
    elif x<=30:
        return 'C'
    else:
        return 'B'
    
final_mean=d1['final'].mean()
final_std=d1['final'].std()
d1['final_grade']=d1['final'].map(final_grade)

d1
```

## 과제2

```python
## 주어진 d1에 even_count 행을 추가하는데
## even_count 행에는 각 열별 짝수의 개수를 넣으시오

d1=pd.read_table('pandas_data3-Copy1.txt', index_col=0)
d1=d1.iloc[:10,1:]
d1

# 코드 작성 시작

# def f2(x):
#     x2=x.loc[x%2==0]
#     return np.sum(x2.index)

# d1.loc['even_count']=d1.apply(f2)
d1.loc['even_count']=d1.apply(lambda col:np.sum(col%2==0),axis=0)

d1
```
