# 문제7

```python
# data1.txt를 읽은 df1에서,
# final 점수가 final평균이상이고, mid 점수가 mid 평균이상인 학생의 데이터만 담은 df2 구하시오

df1=pd.read_table('data1.txt', sep='\t', index_col=0, encoding="cp949")
df1.head()

#=== 코드작성 ===#

final_mean=df1['final'].mean()
mid_mean=df1['mid'].mean()
df2=df1.loc[(df1['final']>=final_mean)&(df1['mid']>=mid_mean)]

print('### df2: final 점수가 final평균이상이고, mid 점수가 mid 평균이상인 학생의 데이터만 담은 dataframe')
df2
```

# 문제8

```python
# data1.txt를 읽은 df1에서,
# pass를 받은 학생의 이름을 구하시오 (조건1과 2를 동시에 만족)
# pass 조건1) final 9등 이하 (1~9등)
# pass 조건2) att 6점이상

df1=pd.read_table('data1.txt', sep='\t', index_col=0, encoding="cp949")
df1.head()

#=== 코드작성 ===#

s1=df1.sort_values('final',ascending=False).iloc[:9]
s1=s1.loc[s1['att']>=6]
name=s1.index

print('### name: pass한 학생의 이름')
name
```

# 문제10

```python
## 주어진 df1에 'A/B' 열을 추가하는데,
## 'A/B'열에는 (mid점수 + 20) > final점수 인 학생은 'A' 그렇지않은 학생은 'B'를 넣으시오

df1=pd.read_table('data1.txt', index_col=0)
df1=df1.iloc[:10,1:]
df1

#=== 코드작성 ===# 

def f1(x):
    global final
    if (x+20)>final:
        return 'A'
    return 'B'

final=df1['final']
df1['A/B']=df1['mid'].apply(f1)

print('### df1: "A/B" 열을 추가한 df1')
df1
```

# 문제11

```python
## 주어진 df1에 각 열별 평균과 중앙값을 담은 두 행을 추가하시오

df1=pd.read_table('data1.txt', index_col=0)
df1=df1.iloc[:10,1:]
df1

#=== 코드작성 ===# 

def f1(col):
    m1=np.mean(col)
    a1=np.median(col)
    return pd.Series([m1,a1], index=['mean','median'])

df2=df1.apply(f1)
df3=pd.concat([df1,df2])

print('### df3: df1에 각 열별 평균과 중앙값을 담은 두 행을 추가한 dataframe')
df3
```

# 문제13

```python
# 주어진 df1에서 'year' 별로 final 점수가 짝수인 점수의 개수를 구하시오

df1=pd.read_table("data1.txt", index_col=0)
df1=df1.head(10)
df1

#=== 코드작성 ===# 

s1=df1.groupby('year')['final'].apply(lambda x:np.sum(x%2==0))

print('### s1: year 별로 final 점수가 짝수인 점수의 개수를 담은 series')
s1
```

# 문제14

```python
# df1에, df2의 phone 정보를 추가한 df3를 구하시오
# (df3에는 df1에 있는 name만 나오도록 하시오)
df1=pd.DataFrame({"name":['a','b','c','d'], 'age':[1,2,3,4]})
df2=pd.DataFrame({"name":['a','b','d','e','f'], 'phone':[9,8,7,6,5]})
df1
df2

#=== 코드작성 ===# 

df3=pd.merge(df1,df2, left_on='name',right_on='name',how='left')

print('### df3: df1의 name에 대하여, df2의 phone 정보를 merge한 dataframe')
df3
```
